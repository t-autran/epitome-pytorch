```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
import sys,os,time
import numpy as np
import rasterio
from rasterio.windows import Window
import matplotlib.pyplot as plt
import skimage.measure
import scipy
from scipy.spatial.distance import euclidean
from scipy.signal import fftconvolve
import pandas as pd
import pickle
import cv2
```

```python
p = pd.read_csv("big_lc_paths.csv", header=None)
paths = p[0].tolist()
ds = 2
save_path = "downsampled_data"+str(ds)+"/"
'''
0       ???    BLACK  UNUSED
1       FOREST GREEN
2       WATER  BLUE
3       FIELD  BROWN
4       OCEAN  ORANGE
5       BARREN YELLOW
6       BUILT  PINK
'''
palette = np.array([[0,0,0],[0,153,51],[0,102,255],[153,102,51],[255,153,0],[255,255,0],[204,102,153]])

def fn(i):
    return paths[i].replace('/','-')[:-7]

def disp(pic):
    plt.figure(figsize=(6,6))
    plt.imshow(pic)
    plt.axis('off')
    plt.show()
    plt.close()

def ds_data(num):    
    if not os.path.exists(save_path):
        os.makedirs(save_path)
    space = 0.
    s = float(time.time())
    for i in range(num):
        if (i%5==4):
            print("Tile %i of %i. Time: %.2f s. Space: %.2f MB" % (i+1,num,float(time.time())-s,space/1024**2))
        lc_tile = rasterio.open('/mnt/blobfuse/resampled-lc/'+paths[i],'r')
        lc = np.squeeze(lc_tile.read())
        lc_tile.close()
        lc[lc==1]=10
        lc[lc==2]=1
        lc[lc==10]=2
        lc[lc==4]=0
        lc[lc==15]=4
        tile = skimage.measure.block_reduce(lc,(ds,ds),np.max)
        path_im = save_path + fn(i)+'.npz'
        np.savez_compressed(path_im,a=tile)
        space += os.stat(path_im).st_size
    fin_time = float(time.time())-s
    fin_size = space/1024**2
    print("Finished! Time: %.2f s. Space: %.2f MB" % (fin_time,fin_size))
    print("Averaged %.3f seconds, %.0f KB per tile" % (fin_time/num,1024*fin_size/num))
```

```python
def channel_conv(big,small):
    channels = np.unique(small)
    total = np.zeros(big.shape,dtype=np.float64)
    for i in channels:
        total += fftconvolve(1*(big==i),1*(small==i)[::-1,::-1],mode='same')
    return np.round(total)

def hist(f):
    temp = np.bincount(f.flatten())[1:]
    return np.pad(temp,(0,6-len(temp)),'constant')

def conv_lookup(query,num_tiles,nns,out_path,to_save):
    time0 = float(time.time())
    size_ds = query.shape[0]
    half = size_ds//2
    results = {}
    for i in num_tiles:
        if (i%10==9):
            print("Searching tile %i..." % (i+1))
        tile = np.load(save_path+fn(i)+'.npz')['a']
        a = channel_conv(tile,query)
        #number of NNs hyperparameter
        for j in range(20):
            dist = np.max(a)
            (z1,z2) = np.unravel_index(np.argmax(a),(a.shape))
            a[z1-half:z1+half,z2-half:z2+half]=0
            z1,z2 = z1-half,z2-half
            results[-dist-np.random.random()/10000] = [i,z1,z2]#,tile[z1:z1+size_ds,z2:z2+size_ds]]
    time1 = float(time.time())-time0
    print("Finished searching in %.1f seconds (averaged %.2f seconds/tile)" % (time1,time1/len(num_tiles)))
    print(len(results.keys()))
    qh = hist(query)
    for key in list(results):
        curr = results[key]
        tile = np.load(save_path+fn(curr[0])+'.npz')['a']
        x_ds, y_ds = curr[1],curr[2]
        curr_tile = tile[x_ds:x_ds+size_ds,y_ds:y_ds+size_ds]
        dist = euclidean(qh,hist(curr_tile))
        #histogram filtering hyperparameter
        if (dist>(size_ds*ds*4.5)):
            del results[key]
        else:
            results[key].append(curr_tile)
    grid_size = min(nns,len(results.keys()))
    #no catch if ALL histograms are filtered
    df = pd.DataFrame()
    cols = int(np.sqrt(grid_size))
    rows = grid_size//cols
    grid_size = rows*cols
    new_cols = cols*2+1
    keys_sorted = sorted(results.keys())
    fig = plt.figure(figsize=(new_cols*8,rows*8))
    all_results, all_results_naip, all_dists = [], [], []
    for i in range(grid_size):
        key = keys_sorted[i]
        curr = results[key]
        acc = np.round(-key*100/size_ds**2,3)
        j = 1+i%cols+new_cols*(i//cols)
        if to_save:
            fig.add_subplot(rows, new_cols, j)
            plt.axis('off')
            plt.imshow(palette[curr[3]])
            plt.title("Tile %i at (%i,%i), %.1f%% accuracy" % (curr[0],curr[1],curr[2],acc))
            with rasterio.open("/mnt/blobfuse/esri-naip/"+fn(curr[0]).replace('-','/')+".mrf") as src:
                naip_tile = src.read([1,2,3],window=Window(curr[2]*ds,curr[1]*ds,size_ds*ds,size_ds*ds))
            naip = np.rollaxis(naip_tile,0,3)
            j += cols+1
            fig.add_subplot(rows, new_cols, j)
            plt.axis('off')
            plt.imshow(naip)
            plt.title("Tile %i at (%i,%i), %.1f%% accuracy" % (curr[0],ds*curr[1],ds*curr[2],acc))
            df = df.append(pd.DataFrame([curr[0],acc]).transpose(),ignore_index=True)
        all_results.append(curr[3])
        #all_results_naip.append(naip)
        all_dists.append(acc)
    if to_save:
        df.to_csv(out_path+"tiles_used.csv",index=False,header=None)
        plt.savefig(out_path+"results.png")
        print("Finished saving %i results in %.1f seconds" %(grid_size,float(time.time())-time0-time1))
    plt.close()

    return all_results, all_dists

def query_many(q,size,num_tiles,nns):
    out_path = 'query_with_ds_'+str(ds)+'/'+str(size)+','+str(q[0])+','+str(q[1])+','+str(q[2])+'/'
    if not os.path.exists(out_path):
        os.makedirs(out_path)
    with rasterio.open("/mnt/blobfuse/esri-naip/"+fn(q[0]).replace('-','/')+".mrf") as src:
        naip_tile = src.read([1,2,3],window=Window(q[2],q[1],size,size))
    naip = np.rollaxis(naip_tile,0,3)
    tile = np.load(save_path+fn(q[0])+'.npz')['a']
    x_ds, y_ds, size_ds = q[1]//ds, q[2]//ds, size//ds
    query = tile[x_ds:x_ds+size_ds,y_ds:y_ds+size_ds]
    fig = plt.figure(figsize=(16,8))
    fig.add_subplot(121)
    plt.title("Tile %i at (%i,%i)" % (q[0],q[1],q[2]))
    plt.axis('off')
    plt.imshow(naip)
    fig.add_subplot(122)
    plt.imshow(palette[query])
    plt.title("Tile %i at (%i,%i)" % (q[0],x_ds,y_ds))
    plt.axis('off')
    plt.savefig(out_path+"query.png")
    plt.close()
    print("Querying tile %i at coordinates (%i,%i)..." % (q[0],q[1],q[2]))
    conv_lookup(query,num_tiles,nns,out_path,True)
    
def query_random(times,query_range,size,tile_nums,nns):
    size_ds = size//ds
    i = 0
    while (i<times):
        query_tile = np.random.randint(0,query_range)
        x, y = np.random.randint(1800), np.random.randint(1500)
        tile = np.load(save_path+fn(query_tile)+'.npz')['a']
        temp_hist = hist(tile[x//ds:x//ds+size//ds,y//ds:y//ds+size//ds])
        if (np.std(temp_hist) > .3*size_ds**2 or np.count_nonzero(temp_hist)<3):
            print("Skipping histogram "+str(temp_hist)+": too boring.")
        else:
            i += 1
            print("Starting lookup (histogram "+str(temp_hist)+")")
            query_many([query_tile,x,y],size,tile_nums,nns)
    print("Finished!")
```

```python
def recreate_patch(patch,stride,results,nns2,weight):
    step = patch.shape[0]//stride
    #patch2 = np.zeros((patch.shape[0],patch.shape[1],3),dtype=np.uint8)
    patch2 = np.zeros((patch.shape[0],patch.shape[1]),dtype=np.uint8)
    #results, _ = conv_lookup(patch,tilelist,nns,'',True)
    for i in range(stride):
        for j in range(stride):
            curr = patch[step*i:step*i+step,step*j:step*j+step]
            probs, corr = query_inhouse(curr,results,nns2,weight)
            #print([i.shape for i in corr])
            for k in range(curr.shape[0]):
                for l in range(curr.shape[1]):
                    #print([temp[k,l] for temp in corr])
                    temp2 = [temp[k,l] for temp in corr]
                    #dot product gives a gradient probablistic label
                    patch2[step*i+k,step*j+l] = np.random.choice(temp2, p=probs)#np.dot(probs,palette[temp2]).astype(int)#
    return patch2

def query_inhouse(subpatch,results,nns2,weight):
    size_ds = subpatch.shape[0]
    half = size_ds//2
    R = {}
    for i in range(len(results)):
        a = channel_conv(results[i],subpatch)
        #print(a)
        #number of NNs hyperparameter
        for j in range(20):
            dist = np.max(a)
            (z1,z2) = np.unravel_index(np.argmax(a),(a.shape))
            a[z1-half:z1+half,z2-half:z2+half]=0
            z1,z2 = max(z1-half,0),max(z2-half,0)
            key = -dist-np.random.random()/10000
            if ([i,z1,z2] not in R.values()):
                R[key] = [i,z1,z2]
    qh = hist(subpatch)
    for key in list(R):
        curr = R[key]
        tile = results[curr[0]]
        x_ds, y_ds = curr[1],curr[2]
        curr_tile = tile[x_ds:x_ds+size_ds,y_ds:y_ds+size_ds]
        #print(qh,hist(curr_tile))
        dist = euclidean(qh,hist(curr_tile))
        #histogram filtering hyperparameter
        if (dist>(size_ds*ds*2.5) or curr_tile.size!=size_ds**2):
            del R[key]
        else:
            R[key].append(curr_tile)
    keys_sorted = sorted(R.keys())
    b, corr = [],[]
    num = min(nns2,len(keys_sorted))
    for i in range(num):
        curr = R[keys_sorted[i]]
        b.append(-keys_sorted[i])
        corr.append(curr[3])
    b = np.array(b)/100
    b = weight*np.max(b)*b
    probs = np.exp(b)/np.sum(np.exp(b))
    return probs,corr

def fillin(patch):
    for i in range(1,patch.shape[0]-1):
        for j in range(1,patch.shape[1]-1):
            val = patch[i,j]
            comps = [patch[i-1,j],patch[i,j-1],patch[i+1,j],patch[i,j+1]]
            if val not in comps:
                patch[i,j] = scipy.stats.mode(comps)[0][0]
    return patch
```
